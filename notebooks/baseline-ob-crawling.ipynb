```python
import requests
import threading

import src
from src.websocket_api import handlers
from src.websocket_api import listeners
import subprocess
import logging
logging.basicConfig(level=logging.INFO)
```

```python
items = list()
for symbol in ['try', 'usdt', 'btc']:
    response = requests.get(f'https://api.btcturk.com/api/v2/ticker/currency?symbol={symbol}')
    for item in response.json()['data']:
        items.append(item['pair'])
```

```python
results = list()
```

```python
def on_message(ws, message):
    result = handlers.handle_channel_message(message)
    global results
    results += [result]
    
def on_error(*args):
    print(args)
    print()

def on_close(*args):
    print(args)
```

```python
threads = list()
```

```python
for item in items:
    _thread = threading.Thread(
        target=listeners.get_listener,
        kwargs={
            'on_message': on_message,
            'on_close': on_close,
            'on_error': on_error,
            'event': item,
            'channel': 'orderbook',
            'time_delay': 1
        }
    )
    threads += [_thread]
```

```python
def thread_control():
    
    while True:
        import time
        global threads
        global results

        counter = 0

        for thread in threads:
            if thread.is_alive():
                continue
            else:
                counter += 1
                thread.start()

        print(f'Started threads {counter}')
        print(f'Total records {len(results)}')

        time.sleep(1)
```

```python
control = threading.Thread(
    target=thread_control,
)
```

```python
control.start()
```
